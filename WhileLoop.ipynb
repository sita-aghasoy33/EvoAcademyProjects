While loops are infinite loops. We do not know the number of steps

```python
i = 0
while i < 5:
   print(i)
    i += 1
```

```python
x = 1
while True:
    print(x)
    if x == 3:
        break
    x += 1
```

```python
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue
    print(n)
```

```python
pwd = ""
while pwd != "1234":
    pwd = input("Enter password: ")
print("Access granted")
```

```python
count = 5
while count > 0:
    print(count)
    count -= 1

print("Done!")
```
